```python
import pdfplumber
import pandas as pd
import logging
logging.getLogger("pdfminer").setLevel(logging.ERROR)
```

```python
pdf_path = "3LAC.pdf"

all_tables = []

with pdfplumber.open(pdf_path) as pdf:
    first_page = pdf.pages[0] 
    first_table = first_page.extract_table()

    if first_table:
        headers = first_table[0] 
        headers = [i.replace('\n', ' ') for i in headers]
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            df_page = pd.DataFrame(table[1:], columns=headers)

            df_page = df_page.map(lambda x: x.replace('\n', ' ') if isinstance(x, str) else x)

            all_tables.append(df_page)
```

```python
all_tables[1]
```

```python
if all_tables:
    df_all = pd.concat(all_tables, ignore_index=True)
    df_all.to_csv("3LAC_combined.csv", index=False)

    print("Table extraction completed successfully!")
    print(df_all.head())
else:
    print("No tables found in the PDF.")
```
